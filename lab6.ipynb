**Question 1
Write a random function call that returns 10 ounces 9 out of 10 times and 9 ounces one out of ten.**

```python
import random

def random_bottle_weight():
    # Create a list representing 10 bottles: 9 are 10oz, 1 is 9oz
    # This gives us a 90% chance of 10oz and 10% chance of 9oz
    possibilities = [10, 10, 10, 10, 10, 10, 10, 10, 10, 9]

    # Pick one randomly
    return random.choice(possibilities)

# Test it once
print(f"One random bottle: {random_bottle_weight()} oz")
```

**Question 2
What is the expected weight of a random pull?**

```python
# Theoretical Calculation
# 90% chance of 10oz, 10% chance of 9oz
expected_weight = (0.9 * 10) + (0.1 * 9)
print(f"The theoretical expected weight is: {expected_weight} oz")
```

Question 3
Test your expected weight by running a simulation drawing at least 100 bottles and taking the average of the weights. Does it agree with your theoretical result?

```python
# We need to sum up the weights of 100 random pulls
total_weight = 0
number_of_pulls = 100

for i in range(number_of_pulls):
    total_weight += random_bottle_weight()

average_weight = total_weight / number_of_pulls

print(f"Average weight after {number_of_pulls} pulls: {average_weight} oz")
# You should see a number very close to 9.9
```

Question 4
We are particularly interested in how many pulls would be required to find the case that was short. If you test one at a time, how many pulls do you expect before you find the case that is short?

```python
# Theoretical Calculation
# The probability (p) is 0.1 (1 in 10).
# The expected number of pulls is 1 / p
expected_pulls = 1 / 0.1
print(f"Theoretical expected pulls to find the short bottle: {expected_pulls}")
```

Question 5
Create a function that creates a sequence of pulling bottles and stops when you find the bottle with 9 ounces. Test this function a bunch of times. Do you notice anything odd when comparing it to your previous result?

```python
def count_pulls_to_find_short():
    count = 0
    while True:
        count += 1
        bottle = random_bottle_weight()
        if bottle == 9:
            return count # Stop and return the number of pulls

# Run this experiment 1000 times to get a good average
total_pulls = 0
experiments = 1000

for i in range(experiments):
    total_pulls += count_pulls_to_find_short()

average_pulls = total_pulls / experiments
print(f"Simulated average pulls to find the short bottle: {average_pulls}")
```

Question 6
Propose a modification to the coding you did in the first step that would improve the results. Hint: Consider a different data structure that would be more applicable to the real world case.

Note: In the real world, when you check a bottle, you don't put it back (sampling without replacement). We should use a list that gets smaller.

```python
def real_world_simulation():
    # Create the case of 100 bottles (90 normal, 10 short)
    # We use a list to represent the physical case
    case_of_bottles = [10]*90 + [9]*10

    # Shuffle them so they are in random positions
    random.shuffle(case_of_bottles)

    pulls = 0
    # Loop until the case is empty
    while len(case_of_bottles) > 0:
        pulls += 1
        # .pop() removes the bottle from the list (takes it out of the case)
        bottle = case_of_bottles.pop()
        if bottle == 9:
            return pulls # Found it!
    return pulls

# Test the real world version
total_real_pulls = 0
for i in range(1000):
    total_real_pulls += real_world_simulation()

print(f"Real world average pulls (without replacement): {total_real_pulls / 1000}")
```

Question 3
Create a randomization and test the previous two questions answers through randomization. Discuss the results.

```python
# Create the stack of plates
# 6 of each color
colors = ['Pink', 'Blue', 'Black', 'Green']
full_stack = []
for color in colors:
    for i in range(6):
        full_stack.append(color)

# Simulation
count_pink = 0
count_at_least_one = 0
trials = 1000

for i in range(trials):
    # Test Q1: Draw 1 plate
    if random.choice(full_stack) == 'Pink':
        count_pink += 1

    # Test Q2: Draw 4 plates (sample means no duplicates)
    hand = random.sample(full_stack, 4)
    if 'Pink' in hand:
        count_at_least_one += 1

print(f"Simulated Prob (1 Pink): {count_pink / trials}")
print(f"Simulated Prob (At least 1 Pink in 4): {count_at_least_one / trials}")
```

Question 4, 5, 6, 7 & 8
# 4. Create a data structure that can account for [Last In First Out]. 5. Add dishwasher logic (shuffle when size > 12). 6. Add rejection process (keep pulling until Pink found). 7. Run for 1000 days. 8. Explore usage counts. **Combined these steps into one complete simulation block below.**

```python
import random

# --- SETUP ---
# Create all 24 plates with unique IDs (e.g., "Pink-1")
all_plates = []
for color in ['Pink', 'Blue', 'Black', 'Green']:
    for i in range(6):
        all_plates.append(f"{color}-{i+1}")

random.shuffle(all_plates)

# Data structures: 'stack' is the cabinet (LIFO), 'dishwasher' holds dirty plates
stack = all_plates[:]
dishwasher = []
usage_counts = {plate: 0 for plate in all_plates}

days = 1000

# --- SIMULATION LOOP ---
for day in range(days):

    # 1. Check if it's time to wash plates (if 12 or more are dirty)
    if len(dishwasher) >= 12:
        random.shuffle(dishwasher)
        stack.extend(dishwasher) # Clean plates go back on top of the stack
        dishwasher = []

    # 2. Set the table (The Rejection Process)
    table_plates = []
    rejected_plates = []

    # First, draw 4 plates for the table
    while len(table_plates) < 4:
        # Emergency wash if stack runs out
        if len(stack) == 0:
            random.shuffle(dishwasher)
            stack.extend(dishwasher)
            dishwasher = []

        plate = stack.pop() # Pull from the top (Last In, First Out)
        table_plates.append(plate)

    # Check for a Pink plate
    has_pink = False
    for plate in table_plates:
        if "Pink" in plate:
            has_pink = True

    # If no pink, continue pulling until one is found
    while not has_pink:
        if len(stack) == 0: # Emergency wash check
            random.shuffle(dishwasher)
            stack.extend(dishwasher)
            dishwasher = []

        new_plate = stack.pop()

        if "Pink" in new_plate:
            # Found pink: swap it with a plate from the table
            rejected_plates.append(table_plates.pop())
            table_plates.append(new_plate)
            has_pink = True
        else:
            # Not pink: put it directly in the reject pile
            rejected_plates.append(new_plate)

    # 3. Clean Up
    stack.extend(rejected_plates) # Rejected plates go back on top

    # Count usage and move used plates to the dishwasher
    for plate in table_plates:
        usage_counts[plate] += 1
        dishwasher.append(plate)

# --- RESULTS (Q8) ---
# Sort the final usage counts to see which plates were used most
sorted_usage = sorted(usage_counts.items(), key=lambda x: x[1], reverse=True)

print("--- Plate Usage after 1000 Days ---")
for plate, count in sorted_usage:
    print(f"{plate}: {count} times used")
```
